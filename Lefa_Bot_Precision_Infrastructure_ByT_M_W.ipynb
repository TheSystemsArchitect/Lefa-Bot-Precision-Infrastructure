```python
import pandas as pd
import numpy as np

# Creating the raw, "messy" Lefa-Bot dataset
data = {
    'timestamp': ['2026-02-22 10:00', '2026-02-22 12:00', None, '2026-02-22 16:00', '2026-02-22 10:00'],
    'sensor_id': ['BOT_01', 'BOT_01', 'BOT_02', 'BOT_01', 'BOT_01'],
    'soil_moisture': [35.5, 34.2, 999, 32.1, 35.5],  # 999 is an outlier
    'thc_level': [0.12, 0.13, 0.15, 0.55, 0.12]     # 0.55 is a compliance risk
}
df = pd.DataFrame(data)
df.to_csv('lefa_bot_raw.csv', index=False)
print("✅ Raw dataset created!")
```

```python
def clean_bot_data(data):
    cleaned_df = data.copy()
    cleaned_df = cleaned_df.drop_duplicates() # Remove duplicate logs
    cleaned_df = cleaned_df.dropna(subset=['timestamp']) # Remove missing times
    # Replace outliers (999) with the average
    mean_val = cleaned_df[cleaned_df['soil_moisture'] <= 100]['soil_moisture'].mean()
    cleaned_df.loc[cleaned_df['soil_moisture'] > 100, 'soil_moisture'] = mean_val
    return cleaned_df

# Execute and view results
final_data = clean_bot_data(df)
print("--- CLEANED DATA RESULTS ---")
print(final_data)
```

```python
import pandas as pd

# Mock dataset representing raw IoT sensor logs
raw_data = {
    'timestamp': ['2026-02-22 10:00', '2026-02-22 10:00', None, '2026-02-22 14:00'],
    'soil_moisture': [35.5, 35.5, 40.0, 999], # Note the duplicate and the 999 outlier
    'uv_index': [5, 5, 6, 11]
}
df = pd.DataFrame(raw_data)

# The Cleaning Function
def clean_lefa_data(data):
    df_clean = data.drop_duplicates()
    df_clean = df_clean.dropna(subset=['timestamp'])
    # Replace the 999 error with the average moisture
    avg_moisture = df_clean[df_clean['soil_moisture'] < 100]['soil_moisture'].mean()
    df_clean.loc[df_clean['soil_moisture'] > 100, 'soil_moisture'] = avg_moisture
    return df_clean

print("--- Cleaned Data Output ---")
print(clean_lefa_data(df))
```
